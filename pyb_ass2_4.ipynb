```python
#Assignment 3
'''
Given a string, check if it is a palindrome

Example:
is_palindrome("radar")
Output: True
is_palindrome("hello")
Output: False
is_palindrome("")
Output: True
Do this in 2 ways:
Iteratively and Recursively
'''
def is_palindrome_iterative(s):
    left = 0
    right = len(s) - 1
    while left < right:
        if s[left] != s[right]:
            return False
        left += 1
        right -= 1
    return True

# Recursive approach
def is_palindrome_recursive(s):
    if len(s) <= 1:
        return True
    if s[0] != s[-1]:
        return False
    return is_palindrome_recursive(s[1:-1])
print(is_palindrome_iterative("radar"))
print(is_palindrome_iterative("hello"))
print(is_palindrome_recursive("radar"))
print(is_palindrome_recursive("hello"))
```

```python
# Iterative approach
def is_palindrome_iterative(s):
    left = 0
    right = len(s) - 1
    while left < right:
        if s[left] != s[right]:
            return False
        left += 1
        right -= 1
    return True

# Recursive approach
def is_palindrome_recursively(s):
    if len(s) <= 1:
        return True
    if s[0] != s[-1]:
        return False
    return is_palindrome_recursively(s[1:-1])

# Test cases
print(is_palindrome_iterative("hello"))
print(is_palindrome_iterative("radar"))
print(is_palindrome_recursively("radar"))
print(is_palindrome_recursively("hello"))

# Test cases
test_cases = [
    ("radar", True),      # Simple palindrome
    ("boommoob", True),   # Simple palindrome
    ("world", False),     # Not a palindrome
    ("a", True),          # Single character
    ("", True),           # Empty string
    ("RadAr", False),     # Case-sensitive test
]

# Iterate over test cases and print results
for i, (input_str, expected) in enumerate(test_cases):
    print(f"Test Case {i + 1}: Input = '{input_str}'")
    result_iterative = is_palindrome_iterative(input_str)
    result_recursive = is_palindrome_recursive(input_str)
    print(f"  Iterative Output: {result_iterative}")
    print(f"  Recursive Output: {result_recursive}")
    print(f"  Expected Output: {expected}")
    print(f"  Iterative Pass: {result_iterative == expected}")
    print(f"  Recursive Pass: {result_recursive == expected}")
    print("-" * 30)
```

```python
#Assignment 4
'''
Given a Number, check if it is a prime number

Example:

is_prime(2)
Output: True

is_prime(6)
Output: False

is_prime(1)
Output: False
'''
#sol
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, n):
        if n % i == 0:
            return False  # divisible → not prime
    return True          # divisible nahi → prime
print(is_prime(2))
print(is_prime(6))
print(is_prime(1))
```

```python
# Function to check prime
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

# Test cases
test_cases = [
    (2, True),        # Smallest prime number
    (3, True),        # Small prime number
    (4, False),       # First composite number
    (17, True),       # Prime number
    (19, True),       # Prime number
    (20, False),      # Composite number
    (1, False),       # Not a prime number
    (0, False),       # Not a prime number
    (-5, False),      # Negative numbers are not prime
    (97, True)        # Large prime number
]

# Iterate through test cases
for i, (input_num, expected) in enumerate(test_cases):
    print(f"Test Case {i + 1}: Input = {input_num}")
    result = is_prime(input_num)
    print(f"  Output: {result}")
    print(f"  Expected Output: {expected}")
    print(f"  Pass: {result == expected}")
    print("-" * 30)
```
