```python
# Assignment 1
'''
Create an array of 18 random elements of shape (3, 6) and add the array [0, 0, 2, 4, 5, 3] to each row.
Then reshape it to a (9, 2) array and take its transpose.
Find the locations where all elements are greater than the mean of the array and print it
'''
import numpy as np
# Random array
random_arr = np.random.rand(3, 6)  # Uniform distribution
print("Random Array:\n", random_arr)
#add array
add_arr=np.array([0,0,2,4,5,3])
added_arr=random_arr+add_arr
print("Added Array:\n",added_arr)
#reshape
reshaped_arr=added_arr.reshape(9,2)
print("Reshaped Array:\n",reshaped_arr)
#transpose
transposed_arr=reshaped_arr.T
print("Transposed Array:\n",transposed_arr)
#find locations where all elements are greater than the mean
mean_val=np.mean(transposed_arr)
print("Mean Value:",mean_val)
locations=np.where(transposed_arr>mean_val)
print("Locations where all elements are greater than the mean:\n",locations)
```

```python
# Assignment 2
'''
Create an array of length 20 that models a poisson distribution and center it around the mean and normalise it by dividing it by the standard deviation
(Centering around some value "a" is subtracting "a" from each element and normalising is to divide it by some factor)
'''
import numpy as np
#Create an array of length 20 with Poisson distribution
poisson_arr=np.random.poisson(lam=10,size=20)
print("Poisson Array:\n",poisson_arr)
#Compute mean and standard deviation
mean_val=np.mean(poisson_arr)
std_val=np.std(poisson_arr)
print("Mean Value:",mean_val)
print("Standard Deviation:",std_val)
#Center the array around the mean
centered_arr=poisson_arr-mean_val
print("Centered Array:\n",centered_arr)
#Normalize the array
normalized_arr=centered_arr/std_val
print("Normalized Array:\n",normalized_arr)


```
